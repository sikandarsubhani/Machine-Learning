# **Single file testing**

```python
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.models import load_model  # Import the load_model function

# Load the trained model
model_path = '/content/skin_disease_detection_model.keras'  # Path to your trained model
model = load_model(model_path)  # Load your trained model here

# Load and preprocess an image for testing
img_path = '/content/download.jpeg'  # Replace with the path to your test image
#img_path = '/content/acne-cystic-30.jpg'  # Replace with the path to your test image
img = image.load_img(img_path, target_size=(150, 150))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0  # Rescale to the same scale used during training

# Perform acne detection on the test image
prediction = model.predict(img_array)
print(prediction)
if prediction >= 0.5:
    result = "Acne"
else:
    result = "Non-Acne"

print(f'The image is classified as: {result} (Confidence: {prediction[0][0]:.2f})')
```

# **Folder Testing**

```python
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.models import load_model
import os

# Load the trained model
model_path = '/content/skin_disease_detection_model.keras'  # Path to your trained model
model = load_model(model_path)

# Define the directory containing test images
test_dir = '/content/test/Acne'  # Replace with the path to your test images

# Loop through the images in the directory
for filename in os.listdir(test_dir):
    if filename.endswith('.jpg'):
        img_path = os.path.join(test_dir, filename)

        # Load and preprocess the image for testing
        img = image.load_img(img_path, target_size=(150, 150))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array /= 255.0  # Rescale to the same scale used during training

        # Perform acne detection on the test image
        prediction = model.predict(img_array)
        print(prediction)
        if prediction >= 0.5:
            result = "Acne"
        else:
            result = "Non-Acne"

        print(f'{filename} is classified as: {result} (Confidence: {prediction[0][0]:.2f})')
```
